##Demonstration - Customizing visualizations using Matplotlib

##Imporing Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.gridspec as gridspec
```

## Load dataset

```python
file_path = "Car_sales.csv"
df = pd.read_csv(file_path)
```

## Selecting a subset

```python
df_sample = df.head(10)
```

## Setting style

```python
plt.style.use('ggplot')
```

## Line Plot with Customizations

```python
plt.figure(figsize=(8, 5))
plt.plot(df_sample["Model"], df_sample["Sales in thousands"], marker='o', linestyle='--', color='purple', linewidth=2, markersize=8)
plt.xticks(rotation=45)
plt.xlabel("Car Model")
plt.ylabel("Sales in Thousands")
plt.title("Customized Line Plot")
plt.show()
```

## Bar Chart with Labels

```python
plt.figure(figsize=(8, 5))
bars = plt.bar(range(len(df_sample)), df_sample["Sales in thousands"], color='blue')

plt.xticks(range(len(df_sample)), df_sample["Model"], rotation=45)  # Correct x-ticks
plt.xlabel("Car Model", fontsize=12, fontweight='bold')
plt.ylabel("Sales in Thousands", fontsize=12, fontweight='bold')
plt.title("Bar Chart with Labels")
```

## Scatterplot with Labels

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["Horsepower"], df["Price in thousands"], color='red')

plt.xlabel("Horsepower", fontsize=12, fontweight='bold')
plt.ylabel("Price in Thousands", fontsize=12, fontweight='bold')
plt.title("Scatterplot: Horsepower vs. Price", fontsize=14)
```

## Using GridSpec

```python
fig = plt.figure(figsize=(10, 6))
gs = gridspec.GridSpec(2, 2, figure=fig)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, :])

ax1.bar(df_sample["Model"], df_sample["Sales in thousands"], color='green')
ax1.set_title("Sales")

ax2.scatter(df_sample["Horsepower"], df_sample["Price in thousands"], color='orange')
ax2.set_title("Horsepower vs. Price")

ax3.hist(df["Price in thousands"].dropna(), bins=10, color="blue", edgecolor="black", alpha=0.7)
ax3.set_title("Histogram: Car Prices")

plt.xticks(rotation=45)
plt.show()
```

## Saving a plot

```python
plt.figure(figsize=(8, 5))
plt.plot(df_sample["Model"], df_sample["Sales in thousands"], marker='s', linestyle='-.', color='navy', linewidth=2)
plt.xlabel("Car Model")
plt.ylabel("Sales in Thousands")
plt.title("Saving Example")
plt.xticks(rotation=45)
plt.savefig("custom_plot.png")  # Saves the plot as an image
```
