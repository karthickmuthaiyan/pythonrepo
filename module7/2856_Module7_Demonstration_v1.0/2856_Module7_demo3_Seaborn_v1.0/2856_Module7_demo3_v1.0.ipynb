##Demonstration - Data visualization using Seaborn

##Importing Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Load dataset

```python
file_path = "Car_sales.csv"
df = pd.read_csv(file_path)
```

## 1. Relational Plot (relplot)

```python
sns.relplot(data=df, x="Price in thousands", y="Sales in thousands", hue="Vehicle type")
plt.show()
```

## 2. Scatter Plot (scatterplot)

```python
sns.scatterplot(data=df, x="Horsepower", y="Sales in thousands", hue="Vehicle type")
plt.show()
```

## 3. Line Plot (lineplot)

```python
sns.lineplot(data=df, x="Engine size", y="Fuel efficiency", hue="Vehicle type")
plt.show()
```

## 4. Bar Plot (barplot)

```python
sns.barplot(data=df, x="Vehicle type", y="Sales in thousands", estimator=np.mean)
plt.show()
```

## 5. Count Plot (countplot)

```python
sns.countplot(data=df, x="Vehicle type")
plt.show()
```

## 6. Box Plot (boxplot)

```python
sns.boxplot(data=df, x="Vehicle type", y="Price in thousands")
plt.show()
```

## 7. Histogram (histplot)

```python
sns.histplot(data=df, x="Horsepower", bins=20, kde=True, hue="Vehicle type")
plt.show()
```

## 8. Pair Plot (pairplot)

```python
sns.pairplot(df, vars=["Price in thousands", "Horsepower", "Fuel efficiency"], hue="Vehicle type")
plt.show()
```

## 9. Heatmap (heatmap)

```python
flights = df.pivot_table(index="Manufacturer", columns="Vehicle type", values="Sales in thousands", aggfunc=np.mean).fillna(0)
sns.heatmap(flights, cmap="coolwarm", annot=True, fmt=".1f")
plt.show()
```

## 10. Correlation Heatmap (correlation heatmap)

```python
corr_matrix = df.select_dtypes(include=["float64", "int64"]).corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.show()
```

## 11. Cluster Map (clustermap)

```python
sns.clustermap(corr_matrix, cmap="coolwarm", standard_scale=1)
plt.show()
```
