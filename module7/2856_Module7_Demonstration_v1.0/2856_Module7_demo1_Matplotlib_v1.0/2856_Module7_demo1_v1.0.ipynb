##Demonstration - Data Visualization using Matplotlib

##Importing Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Load dataset

```python
file_path = "Car_sales.csv"
df = pd.read_csv(file_path)
```

## Selecting a subset for better visualization

```python
df_sample = df.head(10)
print(df_sample)
```

## Line Plot - Sales vs. Model

```python
plt.figure(figsize=(8, 5))
plt.plot(df_sample["Model"], df_sample["Sales in thousands"], marker='o', linestyle='-', color='b')
plt.xticks(rotation=90)
plt.xlabel("Car Model")
plt.ylabel("Sales in Thousands")
plt.title("Line Plot: Sales vs. Model")
plt.legend(["Sales in Thousands"])
plt.show()
```

## Bar Chart - Sales by Manufacturer

```python
plt.figure(figsize=(8, 5))
df_sample.groupby("Manufacturer")["Sales in thousands"].sum().plot(kind='bar')
plt.xlabel("Manufacturer")
plt.ylabel("Total Sales (in thousands)")
plt.title("Bar Chart: Sales by Manufacturer")
plt.xticks(rotation=45)
plt.show()
```

## Horizontal Bar Chart - Sales by Manufacturer

```python
plt.figure(figsize=(8, 5))
print(df_sample.groupby("Manufacturer")["Sales in thousands"].sum())
df_sample.groupby("Manufacturer")["Sales in thousands"].sum().plot(kind="barh", color="m")
plt.xlabel("Total Sales (in thousands)")
plt.ylabel("Manufacturer")
plt.title("Horizontal Bar Chart: Sales by Manufacturer")
plt.show()
```

## Stacked Bar Chart - Sales vs. Engine Size

```python
#print(df_sample.pivot(index="Manufacturer", columns="Model", values="Sles in thousands"))
df_sample.pivot(index="Manufacturer", columns="Model", values="Sales in thousands").plot(kind="bar", stacked=True, figsize=(10, 6))
plt.ylabel("Sales in Thousands")
plt.title("Stacked Bar Chart: Sales vs. Manufacturer")
plt.xticks(rotation=45)
plt.show()
```

## Grouped Bar Chart - Sales vs. Price

```python
print(df_sample[["Model", "Sales in thousands", "Price in thousands"]].set_index("Model"))
df_sample[["Model", "Sales in thousands", "Price in thousands"]].set_index("Model").plot(kind="bar", stacked=False ,figsize=(10, 6))
plt.ylabel("Values")
plt.title("Grouped Bar Chart: Sales vs. Price")
plt.xticks(rotation=45)
plt.show()
```

## Histogram - Distribution of Car Prices

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Price in thousands"].dropna(), bins=20, color="g", edgecolor="black", alpha=0.7)
plt.xlabel("Price in Thousands")
plt.ylabel("Frequency")
plt.title("Histogram: Distribution of Car Prices")
plt.show()
```

## Pie Chart - Vehicle Type Distribution

```python
plt.figure(figsize=(6, 6))
df["Vehicle type"].value_counts().plot(kind="pie", autopct='%1.1f%%', colors=["lightblue", "orange"])
plt.title("Pie Chart: Vehicle Type Distribution")
plt.ylabel("")
plt.show()
```

## Scatterplot - Price vs. Horsepower

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["Horsepower"], df["Price in thousands"], color="r",alpha=0.6)
plt.xlabel("Horsepower")
plt.ylabel("Price in Thousands")
plt.title("Scatterplot: Price vs. Horsepower")
plt.show()
```

## Box Plot - Price Distribution

```python
plt.figure(figsize=(8, 5))
plt.boxplot(df["Price in thousands"].dropna(), vert=False)
plt.xlabel("Price in Thousands")
plt.title("Box Plot: Car Price Distribution")
plt.show()
```
