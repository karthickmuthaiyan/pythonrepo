## Important example to understand variable or attribute scopes

```python
# Scope of Variables 
# Example 4 (Important Concept Application)
global_var=10
class Animal:
    local_var=20
    def method4(self,):
        print('Welcome to Animal Class')
        self.local_var+=1
        print(self.local_var)
Tiger=Animal()
Tiger.method4()
print(global_var) # This is a global variable, so it can be accessed anywhere
#print(local_var)   # This variable is local to the class, so it will throw error
# How to access local_var outside the class?
print(Tiger.local_var)  # This will print 21 as we have modified this for Tiger object.
print(Animal.local_var)  # This will print 20 because the variable has not been modified in class level, it was changed for 
#Tiger object
Animal.local_var=25
print(Tiger.local_var)   # This will print 21 which is specific to Tiger object
print(Animal.local_var)   # This will print 25 which has been set using class name Animal
```

```python
# Example 5 (Important Concept Application)
class Edureka:
    x=10
    def method_1(self,x):
        x=20
        print(x)  # This will print the local variable x which is local to method_1
        print(self.x)  # This will print x which is local to class i.e. 10
        print(id(x))  # It will print the memory address of local x for method.
        print(id(self.x))  # It will print the memory address of local x for class
        x+=1   # We are modifying local x to method_1. To modify x (local to class), we need to write self.x
object1=Edureka()
object1.method_1(20)
print(object1.x)
print(Edureka.x)
object1.x=25
print(Edureka.x)
print(object1.x)
print(id(Edureka.x))
print(id(object1.x))
```

## Public, protected, and private methods

```python
# Example 10
class Employee:
    def __init__(self):  
        self.__empId=101
        self._empName='Smith'
        self.salary=25000
    def public_method(self):
        print('This is a public method')
    def _protected_method(self):
        print('This is a protected method')
    def __private_method(self):
        print('This is a private method')
Employee1=Employee()
Employee1.public_method()
Employee1._protected_method()
Employee1._Employee__private_method()  # Through this way, you can access private methods outside the class. It is similar to private attributes
```

## Multiple constructors

```python
# Example 14

class Date:
    def __init__(self, year, month, day): #year-month-day
        self.year = year
        self.month = month
        self.day = day
    @classmethod
    def dmy(cls, day, month, year): #day-month-year    
        cls.year = year   
        cls.month = month 
        cls.day = day 
        # #order of return should be same as init
        return cls(cls.year, cls.month, cls.day)  
    @classmethod
    def mdy(cls, month, day, year):
        cls.year = year
        cls.month = month
        cls.day = day
        #order of return should be same as init
        return cls(cls.year, cls.month, cls.day)
    
a=Date(2016, 12, 11)
print(a.year) #2016
b=Date.dmy(9, 10, 2015)
print(b.year)
a=Date.mdy(7, 8, 2014)
print(a.year) #2014
```

## Decorators in Python

```python
# Example 15

def method1(arg_func):  # Decorator Definition
    def inner_method():   # This is a wrapper function. Here the argument has been called
        print("Welcome to Edureka") 
        arg_func() # The argument function has been called inside the wrapper function
        print("Execute is Done") 
    return inner_method
def method3(): # This function will be called inside the wrapper
    print("Hello python")
# passing argument function inside the decorator to control its behavior 
argument=method1(method3)
# Calling the function
argument()
```

```python
# Example 16 Decorator
# decorator can also be added to a function as shown below

def func_1(arg_func):
    def func_2(*args,**keyargs):
        print('Welcome to 1')
        returned_argument=arg_func(*args,**keyargs)
        print('Welcome to 2')
        return returned_argument
    return func_2

@func_1    # Decorator has been added to a function
def func_3(x,y): 
    print("Welcome to 3") 
    return x*y 
print(func_3(10,20))
```
