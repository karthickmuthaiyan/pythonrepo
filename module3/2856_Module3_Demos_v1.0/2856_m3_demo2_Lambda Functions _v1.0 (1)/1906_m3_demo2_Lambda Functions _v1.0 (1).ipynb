## Demo 2 - Lambda Function
### The below codes demonstrates the usage of lambda functions

```python
# Example 1  (Method 1)
# A simple lamda function where the values 1 and 2 are added
x=(lambda a,b:a+b)(1,2)
print(x)
```

```python
# Example 2  (Method 2)
# We can integrate the two lines into one statement as well
print((lambda a,b:a+b)(1,2))
```

```python
# Example 3 (Method 3)
# the two variables can be added in the next line too
x=lambda a,b:a+b
x(1,2)
```

```python
# Example 4 (map)
# Here a map object is printed which is later converted to a list and then printed
a=[1,2,3,4]
x=map(lambda elem:elem**2,a)
print(x)   # it prints the map object
print(list(x))  # it prints the list
```

```python
# Example 5
# The Filter function is demonstrated below
a=[1,2,3,4]
y=filter(lambda elem:elem>2,a)
print(y)
print(list(y))
```

```python
# Example 6 
# The Filter function is provided here with a range
b=range(-10,10)
z=filter(lambda elem:elem>0 and elem%2!=0,b)
print(z)
print(list(z))
```

```python
# Example 7 
# A Reduce function is shown below which is used with lambda function

from functools import reduce
product = reduce((lambda x, y: x * y), [1, 2, 3, 4])
print(product)
print(type(product))
```

```python
# Example 8 
# Another example of Reduce function
c=range(10,20)
result=reduce((lambda x,y:x+y),c)
print(result)
```
